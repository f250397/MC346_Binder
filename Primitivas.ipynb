## Primitiva de operação

### Método 1
Uso de lista para guardar valores de cada grau da função e operações realizadas individualmente por elementos em lista.

```guile
(define function-data
  '((0 0 1 1)
    (0 0 1 2)
    (1 1 0 0)
    (1 2 3 0)))

(define (sum func1 func2)
  (let ([a (+ (car func1) (car func2))]
        [b (+ (cadr func1) (cadr func2))]
        [c (+ (cadr (cdr func1)) (cadr (cdr func2)))]
        [d (+ (cadr (cdr (cdr func1))) (cadr (cdr (cdr func2))))])
  `(,a ,b ,c ,d)))
```

```guile
(sum (car function-data) (cadr function-data))
```

### Método 2
Apenas mostragem visual da operação, sem o cálculo de equação resultante.

```guile
(define (operation operador eq1 eq2)
  (cond 
    ((eq? operador 'soma)      (list '+ eq1 eq2))
    ((eq? operador 'subtrai)    (list '- eq1 eq2))
    ((eq? operador 'multiplica) (list '* eq1 eq2))
    ((eq? operador 'divide)   (list '/ eq1 eq2))
    ((eq? operador 'potencia)    (list 'expt eq1 eq2))
    (else (list 'error "Operador desconhecido"))))
```

```guile
(display (operation 'soma 'n 1))
```

```guile
(define equacao-base (operation 'multiplica 3 'n))
(display (operation 'soma equacao-base 5)) ;; Saída: (+ (* 3 n) 5)
(newline)
```

```guile
(define termo1 (operation 'potencia 'x 5))    ;; Resultado: (expt x 5)
(define termo2 (operation 'potencia 'x 2))    ;; Resultado: (expt x 2)
(define termo3 (operation 'multiplica 7 'x))  ;; Resultado: (* 7 x)
;; Combinando tudo com a primitiva 'soma'
(define equacao-parcial (operation 'soma termo1 termo2))
(define equacao-final (operation 'soma equacao-parcial termo3))
(display equacao-final) ;; Saída: ((expt x 5) + (expt x 2)) + (* 7 x))
(newline)
```
